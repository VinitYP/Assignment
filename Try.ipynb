```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
HEADERS = ({'User-Agent':
			'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
			AppleWebKit/537.36 (KHTML, like Gecko) \
			Chrome/90.0.4430.212 Safari/537.36',
			'Accept-Language': 'en-US, en;q=0.5'})
```

```python
def name(soup):
    name = soup.find('span', {'id': 'productTitle'})
    return name.text.strip() if name else ''

def price(soup):
    price = soup.find("span", attrs={'class':'a-offscreen'})
    return price.text.strip() if price else ''

def rating(soup):
    rating = soup.find('span', {'class': 'a-icon-alt'})
    return rating.text.strip() if rating else ''

def reviews(soup):
    review = soup.find('span', {'id': 'acrCustomerReviewText'})
    return review.text.strip() if review else ''

def description(soup):
    description = soup.find('div', {'id': 'productDescription'})
    return description.text.strip() if description else ''

def asin(soup):
    asin = soup.find('th', string=' ASIN ')
    return asin.find_next('td').text.strip() if asin else ''

def manufacturer(soup):
    manufacturer = soup.find('th', string=' Manufacturer ')
    return manufacturer.find_next('td').text.strip() if manufacturer else ''
```

```python
def All_details(url):
    response = requests.get(url, headers=HEADERS)
    soup = BeautifulSoup(response.text, 'html.parser')

    product_details = {
        'Product URL': url,
        'Product Name': name(soup),
        'Product Price': price(soup),
        'Rating': rating(soup),
        'Number of reviews': reviews(soup),
        'Description': description(soup),
        'ASIN': asin(soup),
        'Product Description': description(soup),
        'Manufacturer': manufacturer(soup)
    }

    return product_details
```

```python
def all_pages(url, num_pages):
    all_product_detail = []
    
    for page in range(1, num_pages + 1):
        page_url = url + f'&page={page}'
        response = requests.get(page_url, headers=HEADERS)
        soup = BeautifulSoup(response.content, 'html.parser')
        products = soup.find_all('div', {'data-component-type': 's-search-result'})
        i=1
        for product in products:
            product_url = product.find('a', {'class': 'a-link-normal s-no-outline'}).get('href')
            product_details = All_details('https://amazon.com' + product_url)
            all_product_detail.append(product_details)
            i+=1

    return all_product_detail 
```

```python
if __name__ == '__main__':
    base_url = "https://www.amazon.com/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
    nopage = 6
    nourl = 200

    scraped_data = all_pages(base_url, nopage)
    df = pd.DataFrame(scraped_data)
    df.to_csv('aig.csv', index=False)
```
